```python
import google.generativeai as genai
f = open("/Users/Admin/Internship/RAG App using Langchain/.geminiai_api_key.txt")
key = f.read()

genai.configure(api_key=key)
```

```python
from langchain_text_splitters import NLTKTextSplitter

text_splitter = NLTKTextSplitter(chunk_size=500, chunk_overlap=100)

chunks = text_splitter.split_documents(data)

print(len(chunks))

#print(type(chunks[1]))
```

```python
model=GoogleGenerativeAIEmbeddings(google_api_key=key, model="models/embedding-001")
```

```python
from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(chunks, model, persist_directory="./chroma_db_")
db.persist()
```
